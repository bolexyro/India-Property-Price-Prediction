```python
def split_class_heading(row: pd.DataFrame) -> pd.DataFrame:
    if row["CLASS_HEADING"] != "Owner":
        row["temp_class_label"] = row["CLASS_HEADING"].split(": ")[0]
        row["temp_contact"] = row["CLASS_HEADING"].split(": ")[1]
    else:
        row["temp_class_label"] = row["CLASS_HEADING"]
        row["temp_contact"] = np.nan

    return row
```

```python
from typing import List, Dict
landmarks_categories = []

# SO I WANNA CHECK TO MAKE SUERE THE FORMATTE LANDMARK DETAILS IN NOT A NULL VALUE
formatted_landmark_details_without_nan = df["FORMATTED_LANDMARK_DETAILS"].dropna(
)
for i, landmarks in formatted_landmark_details_without_nan.items():

    formatted_landmark_details: List[Dict] = json.loads(
        formatted_landmark_details_without_nan[i].replace("'", '"'))

    for j in range(len(formatted_landmark_details)):
        if formatted_landmark_details[j]["category"] not in landmarks_categories:
            landmarks_categories.append(
                formatted_landmark_details[j]["category"])

print(landmarks_categories)
```

#### I want to try and create separate columns for all these landmarks with their values as numbers.

```python
for landmarks in landmarks_categories:
    df[landmarks] = 0

formatted_landmark_details_without_nan = df["FORMATTED_LANDMARK_DETAILS"].dropna(
)
for i, landmarks in formatted_landmark_details_without_nan.items():

    formatted_landmark_details: List[Dict] = json.loads(
        formatted_landmark_details_without_nan[i].replace("'", '"'))

    for j in range(len(formatted_landmark_details)):
        no_of_lanmarks = formatted_landmark_details[j]["text"].split(" ")[0]
        df.loc[i, formatted_landmark_details[j]["category"]] = no_of_lanmarks
df.head()
```
